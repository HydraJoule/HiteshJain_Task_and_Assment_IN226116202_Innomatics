```python
# Function to analyze engagement on a social media post
def analyze_post_engagement():
    n = int(input("Enter number of posts: "))
    
    likes_list = []
    
    # Taking likes input dynamically
    for i in range(n):
        likes = int(input(f"Enter likes for post {i+1}: "))
        likes_list.append(likes)

    total_likes = sum(likes_list)

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("\nTotal Likes:", total_likes)
    print("Post Status:", status)


analyze_post_engagement()
```

```python
# Function to analyze rainfall data
def check_rainfall():
    n = int(input("Enter number of rainfall records: "))
    
    rainfall_data = []
    
    for i in range(n):
        rain = float(input(f"Enter rainfall value {i+1}: "))
        rainfall_data.append(rain)

    required_level = float(input("Enter required rainfall level: "))

    average = sum(rainfall_data) / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("\nAverage Rainfall:", average)
    print("Rainfall Status:", status)


check_rainfall()
```

```python
# Function to detect duplicate usernames
def detect_duplicate_accounts():
    n = int(input("Enter number of usernames: "))
    
    usernames = []
    
    for i in range(n):
        name = input(f"Enter username {i+1}: ")
        usernames.append(name)

    if len(usernames) != len(set(usernames)):
        print("\nDuplicate Accounts Found: Yes")
    else:
        print("\nDuplicate Accounts Found: No")


detect_duplicate_accounts()
```

```python
# Function to filter premium crop prices
def filter_premium_crops():
    n = int(input("Enter number of crop prices: "))
    
    prices = []
    
    for i in range(n):
        price = float(input(f"Enter price {i+1}: "))
        prices.append(price)

    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("\nPremium Crops:", premium)


filter_premium_crops()
```

```python
# Function to evaluate application health
def check_system_health():
    errors = int(input("Enter error count: "))

    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("\nError Count:", errors)
    print("System Status:", status)


check_system_health()
```

```python
# Function to validate daily transaction
def check_transaction():
    amount = float(input("Enter transaction amount: "))
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("\nTransaction Amount:", amount)
    print("Transaction Status:", status)


check_transaction()
```

```python
# Function to calculate attendance and check eligibility
def check_attendance():
    total_classes = int(input("Enter total number of classes: "))
    attended = 0

    for i in range(total_classes):
        status = input(f"Enter status for class {i+1} (Present/Absent): ")
        if status.lower() == "present":
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("\nAttendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


check_attendance()
```

```python
# Function to calculate electricity bill
def calculate_electricity_bill():
    units = float(input("Enter number of units consumed: "))
    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("\nTotal Bill:", bill)
    print("Usage Level:", usage)


calculate_electricity_bill()
```
