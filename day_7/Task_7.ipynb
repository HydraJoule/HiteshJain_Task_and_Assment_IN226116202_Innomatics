```python
# Problem 1: Smart Light Controller
# Create a class to control a smart light that can be turned ON or OFF
# and display its status.

class SmartLight:
    
    # Constructor to initialize light name
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Default status
    
    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"
    
    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"
    
    # Method to display current status
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Taking dynamic input from user
light_name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")

# Creating object
light = SmartLight(light_name)

# Performing action
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

# Display result
light.display_status()
```

```python
# Problem 2: Employee ID Card System
# Create a class that stores employee name, ID, and department
# and displays ID card details.

class Employee:
    
    # Constructor to store employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    # Method to display ID card
    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Dynamic input
name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
department = input("Enter Department: ")

# Create object
employee = Employee(name, emp_id, department)

# Display ID card
employee.display_id_card()
```

```python
# Problem 3: Mobile Contact Record
# Create a class to store contact name and phone number
# and display contact information.

class Contact:
    
    # Constructor to store contact details
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    # Method to display contact
    def display_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Dynamic input
name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")

# Create object
contact = Contact(name, phone)

# Display contact
contact.display_contact()
```

```python
# Problem 4: Product Price Tag Generator
# Create a class that stores product name and price
# and prints formatted price tag.

class Product:
    
    # Constructor to store product details
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price
    
    # Method to display price tag
    def display_price_tag(self):
        print("\nProduct:", self.product_name)
        print("Price: â‚¹" + str(self.price))


# Dynamic input
product_name = input("Enter Product Name: ")
price = float(input("Enter Price: "))

# Create object
product = Product(product_name, price)

# Display price tag
product.display_price_tag()
```

```python
# Problem 5: Movie Rating Display System
# Create a class that stores movie name and rating
# and displays movie details.

class Movie:
    
    # Constructor to store movie details
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating
    
    # Method to display movie details
    def display_movie(self):
        print("\nMovie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")


# Dynamic input
movie_name = input("Enter Movie Name: ")
rating = float(input("Enter Rating (out of 5): "))

# Create object
movie = Movie(movie_name, rating)

# Display movie details
movie.display_movie()
```

```python
# Problem 6: Delivery Address Manager
# Create a class that stores customer name and address
# and prints delivery details.

class Delivery:
    
    # Constructor to store delivery details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    # Method to display delivery details
    def display_delivery_details(self):
        print("\nDelivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Dynamic input
customer_name = input("Enter Customer Name: ")
address = input("Enter Delivery Address: ")

# Create object
delivery = Delivery(customer_name, address)

# Display delivery details
delivery.display_delivery_details()
```
