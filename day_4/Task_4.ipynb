```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        return "Recharge amount must be at least ₹50."
    
    if plan not in valid_plans:
        return "Invalid plan selected."
    
    return "Recharge Successful!"

# Retry system
while True:
    amount = int(input("Enter recharge amount: "))
    plan = int(input("Enter plan: "))
    
    result = validate_recharge(amount, plan)
    print(result)
    
    if result == "Recharge Successful!":
        break
    else:
        print("Please try again.\n")
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 20
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    print("Average Marks:", average)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

student_marks = [60, 70, 40, 55, 65]

result = process_result(student_marks)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12
    
    fare = base_fare + (distance * per_km)
    
    if peak.lower() == "yes":
        fare += fare * 0.25
    
    return fare

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")
    
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)
    
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0
    
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

result = check_attendance(attendance)
print("Eligibility:", result)
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False
    
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character)"

password = input("Enter password: ")
print(check_password(password))
```
